## 📊 Histogram – Concept Overview

A histogram is a graphical representation used to understand the distribution of numerical data.

### 🔹 How it works:
- The entire data range is divided into **intervals (bins)**.
- Each bin represents a specific range of values.
- The number of data points falling into each bin is counted (**frequency**).
- Bars are plotted where:
  - **X-axis → bins (value ranges)**
  - **Y-axis → frequency (count of values)**

### 🔹 Key Characteristics:
- Helps visualize the **shape of data distribution** (normal, skewed, uniform, etc.)
- Useful for identifying:
  - Patterns
  - Spread (dispersion)
  - Outliers
- Unlike bar charts, histograms are used for **continuous data**

### 🔹 Example Insight:
- Taller bars → higher concentration of values in that range  
- Wider spread → more variability in data

```python
#histogram
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0) # for same random values after every execution of code
scores= np.random.normal(70, 10, 100) # -> (mean, SD, how many values)

plt.hist(scores,bins=5, color="skyblue", edgecolor="black")
plt.xlabel("Random Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
# custom bins

bins=[0, 30, 60, 90] # 0-30 => 0-inclusive and 30-exclusive
plt.axvline(
    x=42,
    label="passing marks",
    color="red",
    linewidth="2",
    linestyle="--"
)
plt.hist(scores,bins, color="skyblue", edgecolor="black")
plt.xlabel("Random Scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Multiple datasets
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

plt.hist(fraud_transactions, bins=10, label='Fraud', color='red', alpha=0.5, edgecolor="black")
plt.hist(legit_transactions, bins=10, label='Legit', color='green', alpha=0.5, edgecolor="black")

plt.xlabel('Transaction Amount ($)')
plt.ylabel('Frequency')
plt.title('Transaction Amount Distribution: Legitimate vs Fraudulent')
plt.legend()
plt.grid(True)
```
