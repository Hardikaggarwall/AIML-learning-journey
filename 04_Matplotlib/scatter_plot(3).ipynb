```python
import matplotlib.pyplot as plt
```

```python
# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors= ["green" if x<=130 else "red" for x in bp]

plt.scatter(age, bp, color=colors, s=110, marker= "<", alpha=0.3)
#plt.scatter(x-axis, y-axis, color, size, point shape, transperency )
#alpha => 0 to 1 : 0 is invisible, 1 is full opacity
plt.title("BP vs Age")
plt.xlabel("Age")
plt.ylabel("Blood Pressure (BP)")
plt.grid()
plt.show()
```

```python
# c-map in scatter plot
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors= ["green" if x<=130 else "red" for x in bp]

plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp)
#s=bp matlab jitna jayada bp utna badha size
#c=bp is the range of the color variation, where a smaller bp value is for a lower spectrum,
# whereas larger BP values represent a larger spectrum
plt.title("BP vs Age")
plt.xlabel("Age")
plt.ylabel("Blood Pressure (BP)")
plt.grid()
plt.colorbar(label="BP")
plt.show()
```

```python
# annotate
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors= ["green" if x<=130 else "red" for x in bp]

for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]), xytext= (age[i]+1,bp[i]+1))
#plt.annotate(kya likh na hai, kidhar likhna hai (x & y cordinates), kis position pe likh na hai)

plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp), max(bp)+5)

plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp)
plt.title("BP vs Age")
plt.xlabel("Age")
plt.ylabel("Blood Pressure (BP)")
plt.grid()
plt.colorbar(label="BP")
plt.show()
```

```python
# Multiple scatter plots
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (°C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (°C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]

plt.scatter(winter_temp, winter_humidity, label="Winter")
plt.scatter(summer_temp, summer_humidity, label="summer")
plt.legend()
plt.xlabel("Temperature (*c)")
plt.ylabel("Humidity (%)")
plt.show()
```
