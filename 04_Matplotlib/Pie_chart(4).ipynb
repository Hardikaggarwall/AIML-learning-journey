```python
import matplotlib.pyplot as plt
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

plt.style.use("dark_background")
plt.pie(amounts, labels= expenses)
plt.show()
plt.style.available
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
colors= ["orange", "pink", "blue", "yellow", "red"]
explode=[0,0,0,0.3,0]

plt.style.use("default")
plt.pie(amounts, labels= expenses, autopct="%1.1f%%",colors=colors,
       explode=explode,
       startangle=90,
       shadow= True)
plt.show()
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
colors= ["orange", "pink", "blue", "yellow", "red"]

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
       wedgeprops = {
           "edgecolor":"black",
           "linewidth": 3,
           "linestyle": "-.",  #--
           "alpha"    : 0.2,
           "fill"     : False,
           "hatch"    : "/",  # -> /, // , ///  
           "width"    : 0.7,
           "antialiased": True
       }
       )

plt.show()
```
