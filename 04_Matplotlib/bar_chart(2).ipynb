```python
# Bar charts
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M
```

```python
plt.bar(years, oscar_revenue, color="green")
plt.title("years vs revenue of oscar movices")
plt.xlabel("Years")
plt.ylabel("revenue in ($M)")

# Add labels on top of the first bar
#plt.text(years[0], oscar_revenue[0],str(oscar_revenue[0]),ha="center")

# Add labels on top of every bar.
for i in range(len(years)):
    plt.text(years[i], oscar_revenue[i]+10, str(oscar_revenue[i]), ha="center")
    # text(pos of text acc to x-axis, pos of y, text, alignment)


plt.ylim(0, max(oscar_revenue)+200) #lower, upper limit - add space
# We can also increase the x-axis limits. 
plt.show()
```

```python
# comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

# plotting bar  chart

import numpy as np
x = np.arange(len(years))
width = 0.4  # by default width = 0.8

plt.bar(x-width/2, oscar_revenue, width, label="oscar_revenue")
plt.bar(x+width/2, non_oscar_revenue, width, label="non_oscar_revenue")
plt.title("oscar vs non_oscar movies revenue")
plt.xlabel("years")
plt.ylabel("revenue ($M)")
plt.xticks(x, years)
plt.legend()
print(plt.xlim())
plt.show()
    
```

```python
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

import numpy as np
x = np.arange(len(years))
width = 0.25

plt.bar(x-width, oscar_revenue, width)
plt.bar(x, non_oscar_revenue, width)
plt.bar(x+width, oscar_revenue, width)
plt.show()
```

```python
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

import numpy as np
x = np.arange(len(years))
width = 0.2

plt.bar(x-(width+0.1), oscar_revenue, width)
plt.bar(x-(width-0.1), non_oscar_revenue, width)
plt.bar(x+(width-0.1), oscar_revenue, width)
plt.bar(x+(width+0.1), non_oscar_revenue, width)
plt.show()
```
