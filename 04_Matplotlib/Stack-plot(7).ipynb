## 📚 Stack Plot – Concept Overview

A stack plot (also called an area chart) is a graphical representation used to visualize how multiple categories contribute to a total over a period of time or across a sequence of values.

### 🔹 How it works:

* Multiple datasets are plotted as filled areas.
* Each dataset is stacked on top of the previous one.
* The height of the top boundary represents the total value.
* Individual layers show the contribution of each category.

### 🔹 Key Characteristics:

* Displays both individual contributions and overall totals.
* Useful for analyzing trends over time.
* Helps compare how different categories contribute to the whole.
* Provides a clearer view of cumulative data than multiple overlapping line plots.

### 🔹 Example Insight:

* Larger areas → greater contribution to the total.
* Increasing total height → overall growth in values.
* Changes in layer thickness → changing contribution of a category over time.


```python
import matplotlib.pyplot as plt
```

```python
# Stack Plot
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
direct = [50, 60, 70, 80, 90, 100, 110]
organic = [30, 40, 50, 55, 60, 70, 80]
social = [20, 25, 30, 35, 40, 50, 60]

# Single day plot
# plt.pie([direct[0], organic[0], social[0]], labels=["direct", "organic", "social"], autopct="%1.1f%%")

# For multiple days
plt.stackplot(days, direct, organic, social, labels=["Direct","Organic","Social"])
plt.title("Marketing Data for the week")
plt.xlabel("Week")
plt.ylabel("no. of Customer")
plt.legend()
```
