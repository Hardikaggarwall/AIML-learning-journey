## This notebook covers:

- Setting visualization themes using `sns.set_theme()`
- Loading and exploring Seaborn built-in datasets
- Creating relational (scatter) plots with `sns.relplot()`
- Mapping variables using `hue`, `style`, and `size`
- Comparing categories using facet plots with `col`
- Customizing plot appearance with different themes
- Visualizing relationships between numerical and categorical variables

```python
# Seaborn

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set Theme
sns.set_theme() # sets default theme - whitegrid, darkdark, white, dark, ticks
```

```python
# Load-Data
tips = sns.load_dataset("tips")
print(type(tips))
```

```python
tips.head()
```

```python
# Plot
sns.set_theme(style="white")
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",    # for blue and orange color dots 
    style="smoker",  # for Dots and Cross style 
    size="size",     # Size from 1-6
)
```

```python
# Plot
sns.set_theme(style="white")
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",      # 2 graph, for lunch and dinner time 
    hue="smoker",    
    style="smoker",  
    size="size",     
)
```
