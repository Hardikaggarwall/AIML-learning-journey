## This notebook covers:

- Creating **Relational Plots** (`scatterplot` and `lineplot`)
- Creating **Categorical Plots** (`barplot` and `boxplot`)
- Creating **Distribution Plots** using `histplot`
- Creating **Matrix Plots** using `heatmap`
- Loading and visualizing Seaborn built-in datasets
- Customizing plots with `hue`, `bins`, `cmap`, and annotations
- Understanding confidence intervals and data distributions through visualizations

```python
# relationship plot 
#1. scatter plot 
#2. line plot 
```

```python
# Seaborn

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set Theme
sns.set_theme() # sets default theme - whitegrid, darkdark, white, dark, ticks
```

```python
# Relational plots
# scatter plot
sns.get_dataset_names()
tips = sns.load_dataset("tips")
print(tips.head())

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
plt.show()
```

```python
# line plot
flights = sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
plt.show()

# shaded region around the line represents the confidence interval (95%) 
```

```python
# Categorical plot 
# 1. bar plot 
# 2. box plot 
```

```python
# Bar plot
sns.barplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.show()
```

```python
# Box plot
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill"
)
plt.show()
```

```python
# Distribution plot 
# 1. histogram
```

```python
# histogram
penguins = sns.load_dataset("penguins")

print(penguins.head())

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)

plt.show()
```

```python
# Matrix plot 
# 1. HeatMap
```

```python
# Heatmap
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
print(flights_pivot)

sns.heatmap(
    flights_pivot,
    cmap="coolwarm", annot=True, fmt="d"
)
plt.title("Heatmap for Passengers")
plt.tight_layout()
plt.show()
```
