## 🔹 Handling Missing Values

Missing values occur when no data is stored for a particular observation.

### ✅ Common Functions:
- isnull()
- dropna()
- fillna()
- ffill()
- bfill()

```python
import pandas as pd
```

```python
df = pd.read_csv("..//DataSets/raw_data.csv")
df.isnull()  # or we can use .isna()
```

```python
df.isnull().sum()   #gives total number of null values in overall column
```

```python
#drop is used to remove all the rows/columns that contain null values.
df.dropna()  #this is for row 
df.dropna(axis = 1)  # this is for column
```

```python
cleaned_data = df.copy()
mean_age = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(mean_age)
cleaned_data
```

```python
df.ffill()
```

```python
df.bfill()
```

## 🔹 Handling Duplicate Data

Duplicate data refers to repeated rows in a DataFrame.

```python
df.duplicated()
df["country"].duplicated()
df[["country", "gender"]].duplicated()
```

```python
df.drop_duplicates()
#df.drop_duplicates(inplace = True) ....... to do changes directly in the main data .
#we always make a copy and do necessory changes , we avoid directly doing to main data
```

```python
df
```

```python
#data types 
```

```python
df.dtypes
```

```python
df2 = df.copy()
df2["age"] = df2["age"].fillna(0)
df2["age"] = df2["age"].astype("int64")

df2.dtypes


```

## 💡 Notes/Summary

- Use `isnull()` to detect missing values
- Use `dropna()` to remove missing data
- Use `fillna()` to replace missing values
- Use `ffill()` and `bfill()` for sequential data filling
- Use `duplicated()` to identify duplicates
- Use `drop_duplicates()` to remove duplicates
