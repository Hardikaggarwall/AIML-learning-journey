# 📊 Pandas Data Filtering

This notebook covers:
- Filtering using DataFrame conditions
- Filtering using `loc` and `iloc`
- Query-based filtering

```python
import pandas as pd
df = pd.read_csv("../DataSets/globalAirQuality.csv")
```

## 🔹 Filtering using DataFrame Conditions

Filtering allows us to extract rows based on specific conditions.

### ✅ Key Points:
- Conditions return boolean values (`True` / `False`)
- Only rows with `True` are selected

```python
# filtering of data:

#1 using df 
df[df["aqi"] > 100]
df[(df["aqi"] > 100) & (df["temperature"] < 30)]
df[df["aqi"] > 100][["city", "aqi"]] #  method 2 : df.loc[df["aqi"] > 100, ["city", "aqi"]] 
df[(df["aqi"] > 100) & (df["temperature"] > 30)][["city","aqi"]]
```

## 🔹 Filtering using `loc` and `iloc`

### ✅ loc (Label-based filtering)
- Uses column names and conditions
- Preferred for conditional filtering

### ✅ iloc (Position-based filtering)
- Uses index positions
- Not commonly used for condition-based filtering

```python
#2 using loc and iloc

aqi_data = df[(df["aqi"] > 100) & (df["temperature"] > 30)][["city","aqi"]]
aqi_data
aqi_data.iloc[0]
# but if we do .loc[0] it will show error because their is no label "0" 
aqi_data.loc[6]
```

## 🔹 Query-based Filtering

The `query()` method allows filtering using a string expression.

### ✅ Advantages:
- Cleaner and more readable syntax
- Useful for complex conditions

```python
# querry based filtering 

df.query("(aqi > 100) & (temperature >30)")
df.query("aqi > 100 and temperature > 30")[["city","aqi"]]

aqi_data = 100 
df.query("aqi > @aqi_data and temperature > 30")
```

## 🧠 Comparison: Filtering Methods

| Method        | Best Use Case                |
|--------------|-----------------------------|
| df[]         | Simple conditions           |
| loc          | Conditional + column select |
| iloc         | Position-based selection    |
| query()      | Clean and readable filters  |
