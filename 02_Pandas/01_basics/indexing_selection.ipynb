## 🔹 Column Selection

Column selection is used to access one or more columns from a DataFrame.

### ✅ Methods:
- Using column name
- Using list of column names

```python
#1 column selection

df["city"]
df[["city","aqi"]]
```

## 🔹 Row Selection using `loc` and `iloc`

### ✅ loc (Label-based indexing)
- Uses row labels (index names)
- Includes last value

### ✅ iloc (Integer-based indexing)
- Uses position (0,1,2,...)
- Excludes last value (like slicing)

```python
#2 row selection using loc
df.loc[0]
df.loc[0:2]   #starting labelled : ending labelled (both inclusive)
```

```python
#2.1 row selection using iloc

df.iloc[0:2]  # starting index : ending index (ending index exclusive as slicing)
```

## 🔹 Cell Selection using `loc` and `iloc`

Cell selection is used to access a specific value from DataFrame.

### ✅ loc
- Uses row label + column name

### ✅ iloc
- Uses row index + column index

```python
#3 cell selection (row and column)

#3.1 using loc 
df.loc[0,"aqi"]
df.loc[0:2,["city","aqi"]]
```

```python
#3.2 using iloc

df.iloc[0]
df.iloc[0:3,1:5]  # do not write like this df.iloc[0:3 , "city"] because iloc is based on index value 
```

```python
#4 selecting only single value 

df.at[0 , "city"]
df.iat[0, 1]
```

##  Key Differences: loc vs iloc

| Feature        | loc                     | iloc                  |
|----------------|------------------------|----------------------|
| Type           | Label-based            | Position-based       |
| Indexing       | Uses names             | Uses numbers         |
| Slice End      | Inclusive              | Exclusive            |
