# 🔗 Merging & Joining in Pandas

## 📌 Topics Covered

- Inner Join
- Left Join
- Right Join
- Outer Join
- pd.merge()

---

```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv(r"../DataSets/raw_data.csv")
df
```

```python
# Merging & Joining
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 300, 180]
})

Inner_join = pd.merge(df_customers, df_orders, on="customer_id")                  # Inner Join
Left_Join  = pd.merge(df_customers, df_orders, on="customer_id", how="left")      # Left Join
Right_Join = pd.merge(df_customers, df_orders, on="customer_id", how="right")     # Right Join
Outer_Join = pd.merge(df_customers, df_orders, on="customer_id", how="outer")     # Outer Join
print(Inner_join)
print(Left_Join)
print(Right_Join)
print(Outer_Join)
```

```python
# Concatenation
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Adam", "Bob", "Charlie"]
})

df2 = pd.DataFrame({
    "id": [4, 5, 6],
    "name": ["David", "Eva", "Frank"]
})

row = pd.concat([df1, df2])                       # Row wise concatenation
col = pd.concat([df1, df2], axis=1)               # Col wise concatenation
newIndex = pd.concat([df1, df2], ignore_index=True) # Row wise concatenation - new index

print(row, "\n")    
print(col, "\n") 
print(newIndex) 
```

## 💡 Notes/Summary

- `pd.merge()` combines DataFrames using a common key.
- `on` specifies the joining column.
- `Inner Join` → Only matching rows.
- `Left Join` → All rows from left table.
- `Right Join` → All rows from right table.
- `Outer Join` → All rows from both tables.
- Missing matches are represented by `NaN`.
- Default join type is `Inner Join`.
