# 🔗 Concatenation in Pandas

## 📌 Topics Covered

- Row-wise Concatenation
- Column-wise Concatenation
- ignore_index
- pd.concat()

```python
import numpy as np
import pandas as pd 
```

```python
# Concatenation
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Adam", "Bob", "Charlie"]
})

df2 = pd.DataFrame({
    "id": [4, 5, 6],
    "name": ["David", "Eva", "Frank"]
})

row = pd.concat([df1, df2])                       # Row wise concatenation
col = pd.concat([df1, df2], axis=1)               # Col wise concatenation
newIndex = pd.concat([df1, df2], ignore_index=True) # Row wise concatenation - new index

print(row, "\n")    
print(col, "\n") 
print(newIndex) 
```

## 💡 Notes/Summary

- `pd.concat()` combines multiple DataFrames.
- `axis=0` → Row-wise concatenation.
- `axis=1` → Column-wise concatenation.
- `ignore_index=True` → Generates a new continuous index.
- Row-wise concatenation stacks records vertically.
- Column-wise concatenation joins DataFrames horizontally.
- Default concatenation is row-wise (`axis=0`).
