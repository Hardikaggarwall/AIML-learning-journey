## 📌 Topics Covered

- Pivot

---

## 🔹 Melt

`melt()` converts data from **wide format** to **long format**.

### ✅ Common Parameters
- `id_vars`
- `value_vars`
- `var_name`
- `value_name`

## 🔹 Pivot

`pivot()` converts data from **long format** to **wide format**.

### ✅ Common Parameters
- `index`
- `columns`
- `values`

---



```python

import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "Country" : ["USA", "USA", "India", "India"],
    "Year" : [2020, 2021, 2020, 2021],
    "Sales" : [100 , 120, 90 , 110],
    "Profit" : [20 , 25, 18, 22]
})

melted_df = df.melt(
    id_vars=["Country", "Year"],   
    value_vars=["Sales", "Profit"],
    var_name="Metrix",
    value_name="Values"   
)

melted_df.pivot(
    index=["Country", "Year"],
    columns="Metrix",
    values="Values"
)
```
