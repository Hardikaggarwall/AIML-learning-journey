## 🔹 Creating New Features

Feature Engineering is the process of creating new features (columns) from existing data to improve analysis or machine learning models.

### ✅ Using apply()

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"../DataSets/raw_data.csv")

df2 = df.copy()

# method 1 (this is preferable)
df2["tax"] = df2["income"].apply(lambda x : "20%" if x>= 60_000 else "10%")

# method 2 (using normal function)

# def tax_value(x):
#     if x >= 60_000:
#         return "20%"
#     else : 
#         return "10%"

# df2["tax"] = df2["income"].apply(tax_value)

df2
```

### 💡 Notes

- Can use lambda functions.
- Can use normal user-defined functions.
- `apply()` is a Higher Order Function because it accepts another function as an argument.

---

## 🔹 Mapping Values

`map()` is used to transform values according to a dictionary.

```python
gender_map = {"Male": "M", "Female": "F", "Unknown": "U"}
df2["gender"] = df2["gender"].map(gender_map)
df2
```

### 💡 Notes

- Useful for encoding categorical values.
- Dictionary keys are replaced by corresponding values.

---

## 🔹 Creating Columns with assign()

`assign()` creates new columns without modifying the original DataFrame.

```python
df2 = df2.assign(new_income = df2["income"]*1.1)
df2
```

## 🔹 Replacing Values

`replace()` replaces specific values in a DataFrame.

```python
df2["country"] = df2["country"].replace("USA", "US")
df2
```

## 🔹 Renaming Labels

### Rename Columns

```python
df2.rename(columns = {"income": "salary" , "gender": "sex"} )
```

### Rename Index

```python
df2.rename(index = {1:"first"})
```

### 💡 Notes

- Does not modify the original DataFrame unless `inplace=True`.

## 🔹 Sorting Data

### Sort by Values

```python
sorted_data = df2.sort_values("income")
sorted_data
```

### Sort by Index

```python
sorted_data.sort_index()
```

### 💡 Notes

- Default sorting order is ascending.
- Use `ascending=False` for descending order.

## 🔹 Write data to csv

```python
df2 = df.copy()

df2 = df2.drop_duplicates()
df2 = df2.fillna(0)
df2 = df2.sort_values("income")
df2 = df2.reset_index(drop=True)
df2.to_csv("sorted_data.csv")
df2
```

## 🔹 GroupBy Aggregation

`groupby()` is used to split data into groups and perform calculations.

```python
df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()

df.groupby("gender")["income"].max()
```

## 🔹 Multiple Aggregations

### Using aggregate()

```python
df.groupby("country")["income"].aggregate(["mean", "max", "min"])
df.groupby("country")["income"].aggregate(avg_income= "mean", max_income = "max", min_income= "min")
```

## 🔹 Aggregating Multiple Columns

### Dictionary Style

```python
df.groupby("country").agg({
    "income": "max",
    "age": "min"
})
```

### Named Aggregation Style

```python
df.groupby("country").agg(
    max_income = ("income", "max"),
    min_age= ("age", "min")
)
```

## 💡 Notes/Summary

- `apply()` → Apply custom functions to data.
- `map()` → Replace values using a dictionary.
- `assign()` → Create new columns.
- `replace()` → Standardize or modify values.
- `rename()` → Change column or index names.
- `sort_values()` → Sort rows using column values.
- `sort_index()` → Sort using index.
- `reset_index()` → Create a fresh index.
- `rank()` → Assign rankings.
- `groupby()` → Split data into groups.
- `agg()` / `aggregate()` → Perform multiple aggregations.
